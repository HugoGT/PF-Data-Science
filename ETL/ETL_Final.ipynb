```python
import random

import pandas as pd
import unidecode
```

```python
df_japon = pd.read_csv('../data/sismos_japon_con_ubicacion.csv')

df_peru= pd.read_csv('../data/sismos_peru_con_ubicacion.csv')

df_usa = pd.read_csv('../data/sismos_EEUU_con_ubicacion.csv')
```

```python
df_japon.head(1)
```

```python
df_peru.head(1)
```

```python
df_peru.drop(columns=("ciudad"), inplace=True)
```

```python
df_usa.tail(1)
```

```python
df_final = pd.concat([df_peru, df_japon, df_usa], ignore_index=True)
```

```python
df_final.info()
```

```python
df_final.insert(0, 'id_sismo', df_final.index+1)
```

```python
df_final.head()
```

```python
df_tsunamis = pd.read_csv('../data/tsunamis.csv')
df_tsunamis['id_sismo'] = df_tsunamis['id_sismo'].astype(int)
df_tsunamis.rename(columns={"max_water_height": "altura_de_ola"}, inplace=True)
df_tsunamis.head()
```

Se rellenan datos faltantes con una regresión

```python
df_daños = pd.read_csv('../data/daños_regresion.csv')
df_daños.head()
```

```python
departamentos_peru = [
    'Amazonas',
    'Ancash',
    'Apurimac',
    'Arequipa',
    'Ayacucho',
    'Cajamarca',
    'Callao',
    'Cusco',
    'Huancavelica',
    'Huanuco',
    'Ica',
    'Junin',
    'La Libertad',
    'Lambayeque',
    'Lima',
    'Loreto',
    'Madre de Dios',
    'Moquegua',
    'Pasco',
    'Piura',
    'Puno',
    'San Martin',
    'Tacna',
    'Tumbes',
    'Ucayali'
]
```

```python
df_daños['pais'] = df_daños['pais'].fillna('Perú')
df_daños['estado'] = df_daños['estado'].fillna(random.choice(departamentos_peru))
```

Estandarizamos columnas estado y pais

```python
def procesar_columnas(df):
    df['estado'] = df['estado'].str.lower().apply(unidecode.unidecode).str.title()
    df['pais'] = df['pais'].str.lower().apply(unidecode.unidecode).str.title()
```

```python
procesar_columnas(df_daños)
procesar_columnas(df_tsunamis)
procesar_columnas(df_final)
```

```python
df_final[df_final['estado'] == 'Mar De Usa'].head()
```

```python
df_final['estado'] = df_final['estado'].replace('Mar De Usa', 'Mar de EEUU').replace('Mar De Eeuu', 'Mar de EEUU')
df_final
```

Creación tabla de dimensión del lugar

```python
# Obtener las combinaciones únicas de las columnas "pais" y "estado"
combinaciones_sismos = df_final[['pais', 'estado']].drop_duplicates()
combinaciones_tsunamis = df_tsunamis[['pais', 'estado']].drop_duplicates()
combinaciones_daños = df_daños[['pais', 'estado']].drop_duplicates()
```

```python
# Crear el nuevo DataFrame con las combinaciones únicas
combinaciones_unicas = pd.concat([combinaciones_sismos, combinaciones_tsunamis, combinaciones_daños], ignore_index=True)
combinaciones_unicas
```

```python
lugar = combinaciones_unicas.drop_duplicates(inplace=True)
lugar = combinaciones_unicas.reset_index(drop=True)
lugar.insert(0, 'id_lugar', range(1, len(lugar) + 1))
```

```python
lugar
```

```python
df_sismos_final = pd.merge(df_final, lugar, on=['pais', 'estado'], how='left')
```

```python
df_sismos_final.drop(columns=["pais", "estado"], inplace= True)
df_sismos_final
```

```python
df_tsunamis_final = pd.merge(df_tsunamis, lugar, on=['pais', 'estado'], how='left')
df_tsunamis_final.drop(columns=["pais", "estado"], inplace= True)
df_tsunamis_final
```

```python
df_daños_final = pd.merge(df_daños, lugar, on=['pais', 'estado'], how='left')
df_daños_final.drop(columns=["pais", "estado"], inplace= True)
df_daños_final
```

```python
df_daños_final.fillna(0, inplace=True)
```

```python
lugar.to_csv("../data/tabla_lugar.csv", index= False)
df_sismos_final.to_csv('../data/tabla_sismos.csv', index=False)
df_daños_final.to_csv('../data/tabla_daños.csv', index=False)
df_tsunamis_final.to_csv('../data/tabla_tsunamis.csv', index=False)
```
