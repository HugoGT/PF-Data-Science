```python
columnas = ['Year', 'Mo', 'Dy', 'Hr', 'Mn', 'Sec','Earthquake Magnitude', 'Country',
            'Maximum Water Height (m)']
origen_tsunamis = pd.concat([origen_peru[columnas], origen_japon[columnas], origen_usa[columnas]])
```

```python
origen_tsunamis.head(12)
```

```python
#Borro la fila de cada columna que venía vacía
origen_tsunamis.dropna(subset=["Year"], inplace= True)
#Reestablesco el index
origen_tsunamis.reset_index(drop=True, inplace=True)
```

```python
#Relleno los nulos de segundos por 00 para poder convertir los datos a enteros
origen_tsunamis["Sec"].fillna("00", inplace=True)
```

```python
#Convierto a enteros los datos
origen_tsunamis[['Year', 'Mo', 'Dy', 'Hr', 'Mn', 'Sec']] = origen_tsunamis[['Year', 'Mo', 'Dy', 'Hr', 'Mn', 'Sec']].astype(int)
```

```python
#Creo columna fecha y la lleno con los datos de la fecha, le cambio a formato fecha
origen_tsunamis['Fecha'] = origen_tsunamis[['Year', 'Mo', 'Dy']].apply(lambda x: '-'.join(x.astype(str)), axis=1)
origen_tsunamis["Fecha"] = pd.to_datetime(origen_tsunamis["Fecha"]).dt.date
```

```python
#Creo columna hora y la lleno con los datos de la hora, le cambio a formato fecha
origen_tsunamis['Hora'] = origen_tsunamis[['Hr', 'Mn', 'Sec']].apply(lambda x: ':'.join(x.astype(str)), axis=1)
origen_tsunamis["Hora"] = pd.to_datetime(origen_tsunamis["Hora"], format='%H:%M:%S').dt.time
```

```python
origen_tsunamis.drop(['Year', 'Mo', 'Dy', 'Hr', 'Mn', 'Sec'], axis=1, inplace=True)
```

```python
# Agregar la columna 'id_tsunami' con números incrementales
origen_tsunamis['id_tsunami'] = pd.Series(range(1, len(origen_tsunamis) + 1))
```

```python
columnas = ['Year', 'Mo', 'Dy', 'Hr', 'Mn', 'Sec','Earthquake Magnitude', 'Country',
            'Area', 'Max Water Height (m)']
efectos_tsunamis = pd.concat([efecto_peru[columnas], efecto_japon[columnas], efecto_usa[columnas]])
```

```python
efectos_tsunamis
```

```python
#Borro la fila de cada columna que venía vacía
efectos_tsunamis.dropna(subset=["Year"], inplace= True)
#Reestablesco el index
efectos_tsunamis.reset_index(drop=True, inplace=True)
```

```python
#Relleno los nulos de segundos por 00 para poder convertir los datos a enteros
efectos_tsunamis["Sec"].fillna("00", inplace=True)
```

```python
#Convierto a enteros los datos
efectos_tsunamis[['Year', 'Mo', 'Dy', 'Hr', 'Mn', 'Sec']] = efectos_tsunamis[['Year', 'Mo', 'Dy', 'Hr', 'Mn', 'Sec']].astype(int)
```

```python
#Creo columna fecha y la lleno con los datos de la fecha, le cambio a formato fecha
efectos_tsunamis['Fecha'] = efectos_tsunamis[['Year', 'Mo', 'Dy']].apply(lambda x: '-'.join(x.astype(str)), axis=1)
efectos_tsunamis["Fecha"] = pd.to_datetime(efectos_tsunamis["Fecha"]).dt.date
```

```python
#Creo columna hora y la lleno con los datos de la hora, le cambio a formato fecha
efectos_tsunamis['Hora'] = efectos_tsunamis[['Hr', 'Mn', 'Sec']].apply(lambda x: ':'.join(x.astype(str)), axis=1)
efectos_tsunamis["Hora"] = pd.to_datetime(efectos_tsunamis["Hora"], format='%H:%M:%S').dt.time
```

```python
efectos_tsunamis.drop(['Year', 'Mo', 'Dy', 'Hr', 'Mn', 'Sec'], axis=1, inplace=True)
```

```python
#Cambio el nombre de la columna para que coincida con la de sismos
efectos_tsunamis.rename(columns={"Area":"Estado"}, inplace= True)
```

```python
#Dejo un solo nombre para los tsunamis de usa
efectos_tsunamis["Country"] = efectos_tsunamis["Country"].replace("USA TERRITORY", "USA")
```

```python
efectos_tsunamis
```

```python
efectos_tsunamis['Fecha'] = pd.to_datetime(efectos_tsunamis['Fecha'])
origen_tsunamis['Fecha'] = pd.to_datetime(origen_tsunamis['Fecha'])
```
