```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
usa = pd.read_csv('../data/sismos_EEUU.csv')
```

```python
usa
```

```python
geolocator = Nominatim(user_agent="usa_test", timeout=6)
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=2)
```

```python
lald= "37.301700,-104.447300"
lald
```

```python
localizacion = geolocator.reverse(lald)
localizacion
```

```python
def combinar_coordenadas(row):
    latitud = row['latitude']
    longitud = row['longitude']
    coordenadas = f'{latitud},{longitud}'

    return coordenadas

# Aplica la función a cada fila del DataFrame y crea la nueva columna
usa['coordenadas'] = usa.apply(combinar_coordenadas, axis=1)
```

```python
coordenadas = usa['coordenadas'].to_list()
coordenadas
```

```python
ubicacion = []
for u in coordenadas:
    ubicacion.append(geolocator.reverse(u))
```

```python
ubicacion
```

```python
country = []
condados = [] #deptos
state = [] #provincias
city = []

for item in ubicacion:
    if item == None:
        country.append('United States')
        condados.append(None)
        state.append('Mar de USA')
        city.append(None)
    else:
        address = item.raw['address']

        country.append(address.get('country'))
        state.append(address.get('state'))
        condados.append(address.get('region'))
        city.append(address.get('city', address.get('village', address.get('town'))))

print(len(country), len(condados), len(state), len(city))
```

```python
usa['country'] = country
usa['state'] = state
usa['city'] = city
```
