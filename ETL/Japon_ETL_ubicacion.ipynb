## Sacando Ubicaciones de Japón

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/sismos_japon_completo.csv')
```

```python
df.info()
```

```python
df['Epicentro'] = df['Epicentro'].apply(lambda x: x.replace(':', '').replace('Cerca de ', ''))
```

```python
# Sacamos prefecturas de Japón
prefecturas_japon = [
    'Hokkaido',
    'Aomori',
    'Iwate',
    'Miyagi',
    'Akita',
    'Yamagata',
    'Fukushima',
    'Ibaraki',
    'Tochigi',
    'Gunma',
    'Saitama',
    'Chiba',
    'Tokio',
    'Kanagawa',
    'Niigata',
    'Toyama',
    'Ishikawa',
    'Fukui',
    'Yamanashi',
    'Nagano',
    'Gifu',
    'Shizuoka',
    'Aichi',
    'Mie',
    'Shiga',
    'Kyoto',
    'Osaka',
    'Hyogo',
    'Nara',
    'Wakayama',
    'Tottori',
    'Shimane',
    'Okayama',
    'Hiroshima',
    'Yamaguchi',
    'Tokushima',
    'Kagawa',
    'Ehime',
    'Kochi',
    'Fukuoka',
    'Saga',
    'Nagasaki',
    'Kumamoto',
    'Oita',
    'Miyazaki',
    'Kagoshima',
    'Okinawa'
]
```

```python
def ubicar_estado(row):
    if 'costa' in row:
        return 'Mar de Japón'
    else:
        prefectura = ''
        for p in prefecturas_japon:
            if p in row:
                prefectura = p

        if prefectura == '':
            return 'Mar de Japón'
        else:
            return prefectura
```

```python
df['pais'] = 'Japón'
df['estado'] = df['Epicentro'].apply(ubicar_estado)
df
```

```python
df = df.rename(columns={
    'Fecha': 'fecha',
    'Hora': 'hora',
    'Latitud': 'latitud',
    'Longitud': 'longitud',
    'Profundidad': 'profundidad',
    'Mag': 'magnitud',
})
df
```

```python
df = df.drop(axis=1, columns=['Epicentro', 'Máxima intensidad sísmica'])
df
```

```python
df.to_csv('../data/sismos_japon_con_ubicacion.csv', index=False)
```
